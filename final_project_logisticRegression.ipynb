# DATA PREPROCESSING

## DATA CLEANING 

```python
data.drop(["Unnamed: 32","id"],axis=1,inplace=True)
data.head()
```

## DATA MODIFICATION

```python
# Convert categorical diagnosis labels into numeric form for machine learning:
# Malignant (M) → 1 and Benign (B) → 0.
# This binary encoding is required because ML models operate on numerical data
# and it also makes the target variable suitable for binary classification.
data['diagnosis'] = data['diagnosis'].map({'M': 1, 'B': 0})
```

```python
data
```

```python
# Convert the 'diagnosis' column to a categorical data type to optimize memory usage
# and clearly indicate that it represents discrete class labels (not numerical values)
data["diagnosis"] = data["diagnosis"].astype("category", copy=False)
data.info()
```

```python
data["diagnosis"].value_counts().plot(kind="bar") #visualization of distribution of data
```

```python
#divide into target variables and predictors
y=data.diagnosis
X=data.drop(["diagnosis"],axis=1)
```

## NORMALIZATION(SCALING)

```python
from sklearn.preprocessing import StandardScaler
# Normalization is essential because many ML algorithms (e.g., SVM, KNN, Logistic Regression)
# are sensitive to feature scale—standardizing ensures all features contribute equally,
# improves convergence speed, and prevents features with larger magnitudes from dominating the model


#create a scalar object
scaler=StandardScaler()

#fit the scaler to the data and transform the data
X_scaled=scaler.fit_transform(X)
```

# TEST-TRAIN SPLIT

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.3,random_state=42)
# Split the dataset into training and testing sets
# This helps evaluate the model on unseen data and check its generalization ability.
# 30% of the data is reserved for testing, while 70% is used for training.
```

# INITIALIZING AND TRAINING MODEL

```python
from sklearn.linear_model import LogisticRegression
```

```python
#creating instance(object of class) of logistic regression  model
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
# Fit (train) the Logistic Regression model on the training data
# Step 1: The model takes a weighted sum of input features:
#         z = b0 + b1*x1 + b2*x2 + ... + bn*xn
# Step 2: This value is passed through the sigmoid function:
#         sigmoid(z) = 1 / (1 + e^(-z))
#         → converts any number into a probability between 0 and 1
# Step 3: If probability is close to 1 → class 1, close to 0 → class 0
# Step 4: The difference between predicted and actual values is measured (error)
# Step 5: Weights are adjusted to reduce this error
# This process repeats until the model learns the best weights
```

# PREDICTING

```python
y_pred = lr.predict(X_test) 
```

# PERFORMANCE OF CLASSIFICATION MODEL AT DIFFERENT THRESHOLDS

## RECEIVER OPERATING CHARACTERISTIC CURVE , AUC

```python
# Significance of the ROC curve:
# - Shows how well the model distinguishes between two classes across all thresholds
# - Visualizes the trade-off between True Positive Rate and False Positive Rate
# - A curve closer to the top-left corner indicates better classification performance
# - Helps compare different models independent of a fixed decision threshold
# - Useful even when the dataset is imbalanced
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Get predicted probabilities for the positive class (class = 1)
# Instead of hard class labels (0 or 1), ROC uses probabilities
y_prob = lr.predict_proba(X_test)[:, 1]

# Compute ROC curve values
# fpr (False Positive Rate): proportion of negative cases wrongly classified as positive
# tpr (True Positive Rate): proportion of positive cases correctly classified
# thresholds: different probability cut-offs used to compute fpr and tpr
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# Compute AUC (Area Under the Curve)
# AUC summarizes the ROC curve into a single number
# It represents the probability that the model ranks a random positive
# example higher than a random negative one
roc_auc = auc(fpr, tpr)

# Plot
plt.figure()
plt.plot(fpr, tpr, label=f"ROC curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], linestyle="--", label="Random")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

## VARIATION OF TPR AND FPR ON CHANGING THRESHOLD VALUE

```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
# - Shows how model performance changes as the classification threshold varies from 0 to 1
# - Helps understand the trade-off between True Positive Rate (recall) and False Positive Rate
# - Lower thresholds increase recall but may raise false positives
# - Higher thresholds reduce false positives but may miss true positives
# - Useful for selecting an optimal threshold based on problem needs
thresholds = np.arange(0, 1.01, 0.05)

rows = []

for t in thresholds:
    # Apply threshold
    y_pred_thresh = (y_prob >= t).astype(int)

    # Confusion matrix
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred_thresh).ravel()

    # Rates
    tpr = tp / (tp + fn) if (tp + fn) != 0 else 0  # Recall
    fpr = fp / (fp + tn) if (fp + tn) != 0 else 0

    rows.append({
        "Threshold": round(t, 2),
        "TPR (Recall)": round(tpr, 3),
        "FPR": round(fpr, 3)
    })

# Create table
threshold_table = pd.DataFrame(rows)

threshold_table
```
